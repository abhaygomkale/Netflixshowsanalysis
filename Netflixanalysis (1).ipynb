```python
titleperdirector=nf.groupby('director')[['title']].count().sort_values(by='title',ascending=False)
```

```python
titleperdirector=titleperdirector.reset_index()
```

```python
titleperdirector.columns
```

```python
top15drc=titleperdirector.head(15)
```

```python
colors = ['green', 'red', 'magenta', 'blue', 'pink', 
          'orange', 'purple', 'cyan', 'yellow', 'lime', 
          'brown', 'navy', 'teal', 'gold', 'gray']
plt.figure(figsize=(10, 5)) 
bars = plt.bar(top15drc['director'], top15drc['title'], color=colors)
plt.title('Top 15 Directors by Number of works')
plt.xlabel('Director Name')
plt.ylabel('No.of Works')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('15topdirectbar.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
show_type=nf.groupby('type')[['show_id']].count()
```

```python
show_type=show_type.reset_index()
```

```python
show_type
```

```python
plt.figure(figsize=(10, 5))
plt.pie(show_type['show_id'],labels=show_type['type'],autopct='%1.f%%',colors=['lightcoral', 'skyblue'])
plt.title('Distribution of Show Types')
plt.axis('equal')
plt.savefig('Distribmvitvpie.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 5))  

plt.hist(nf['release_year'], bins=50, color='skyblue', edgecolor='black')  
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Shows', fontsize=12)
plt.title('Number of Shows Released per Year', fontsize=15)
plt.xticks(rotation=45)  
plt.tight_layout()
plt.savefig('showsperyearhist.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
movie=nf[nf['type']=='Movie']
```

```python
tvshow=nf[nf['type']=='TV Show']
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5),sharey=True)

ax[0].hist(movie['release_year'],bins=20,color='c',edgecolor='Black')
ax[0].set_title('Movies per year')
ax[0].set_xlabel('Year',fontsize=12)
ax[0].set_ylabel('Number of movies',fontsize=12)

ax[1].hist(tvshow['release_year'],bins=20,color='m',edgecolor='Black')
ax[1].set_title('TV Shows per year')
ax[1].set_xlabel('Year',fontsize=12)
ax[1].set_ylabel('Number of TV Shows',fontsize=12)

plt.suptitle('Release Year Distribution by Type', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('moviesshowyearhist.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
moviecountry=nf.groupby('country')['show_id'].count().sort_values(ascending=False).head(10)
```

```python
moviecountry=moviecountry.reset_index()
```

```python
moviecountry.columns
```

```python
moviecountry.rename(columns={'show_id':'No. of movies'},inplace=True)
```

```python
moviecountry.sort_values(by='No. of movies',inplace=True)
```

```python
moviecountry.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(moviecountry['country'], moviecountry['No. of movies'], color='skyblue', edgecolor='black')
plt.title('Countries with most movie count', fontsize=15)
plt.xlabel('No. of movies', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Add data points to the bars
for index, value in enumerate(moviecountry['No. of movies']):
    plt.text(value + 1, index, str(value), va='center', fontsize=10)  # Add some spacing with +1

plt.tight_layout()
plt.grid(linestyle=':', alpha=0.6)
plt.savefig('countmpviebar.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
